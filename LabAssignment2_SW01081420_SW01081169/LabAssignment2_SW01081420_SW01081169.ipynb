```python
# Load the dataset (assuming it's stored as a CSV file)
df = pd.read_csv('Reviews.csv')
```

```python
# Specify the number of rows to keep
num_rows_to_keep = 1000  # Example: Keep 10,000 rows
# Randomly sample the dataset to keep the desired number of rows
df_sampled = df.sample(n=num_rows_to_keep, random_state=42)
```

```python
# check no. of rows & columns
df_sampled.shape
```

```python
df_sampled.head()
```

```python
def label_sentiment(score):
    if score <= 2:
        return 'negative'
    elif score == 3:
        return 'neutral'
    else:
        return 'positive'

# Apply sentiment labeling
df_sampled['Sentiment'] = df_sampled['Score'].apply(label_sentiment)
```

```python
from textblob import TextBlob

# Define a function to get sentiment polarity using TextBlob
def get_textblob_sentiment(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return 'positive'
    elif sentiment_score < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis using TextBlob to the sampled dataset
df_sampled['TextBlobSentiment'] = df_sampled['Text'].apply(get_textblob_sentiment)
```

```python
import nltk 
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a function to get sentiment using VADER
def get_vader_sentiment(text):
    sentiment_scores = sid.polarity_scores(text)
    if sentiment_scores['compound'] >= 0.05:
        return 'positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis using VADER to the sampled dataset
df_sampled['VADERSentiment'] = df_sampled['Text'].apply(get_vader_sentiment)
```

```python
from sklearn.metrics import classification_report
from tabulate import tabulate
```

```python
# Actual sentiment labels based on review scores
actual_sentiment = df_sampled['Score'].apply(lambda score: 'positive' if score > 3 else ('negative' if score < 3 else 'neutral'))

# Classification report for TextBlob sentiment analysis
print("Classification Report for TextBlob Sentiment Analysis:")
print(classification_report(actual_sentiment, df_sampled['TextBlobSentiment']))

# Classification report for VADER sentiment analysis
print("Classification Report for VADER Sentiment Analysis:")
print(classification_report(actual_sentiment, df_sampled['VADERSentiment']))

```
